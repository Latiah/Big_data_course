## Simulate gender

```python
np.random.seed(0)
gender = np.random.choice(["Male", "Female"], 20, p=[0.4, 0.6]) # 40% of male and 60% of Female
gender
```

```python
len(gender)
```

```python
len(gender[gender == "Female"])
```

```python
sum(gender == "Female")
```

```python
print(sum((gender == "Female")))
```

```python
print(sum((gender == "Male")))
```

```python
print(np.sum(gender == "Female"))
```

```python
print(np.mean(gender == "Female"))
```

```python
print(np.mean(gender == "Female"))
```

```python
import pandas as pd

my_dict ={"Age": age,
          "Gender": gender}
data_sim = pd.DataFrame(my_dict)
data_sim
```

```python
data_sim.head()
```

```python
pd.DataFrame(np.array([age, gender]).T, columns = ["Age", "Gender"])
# .T is transpose
```

```python
%pip install matplotlib
```

```python
data_sim.groupby(['Gender'])["Age"].mean().plot(kind ="pie")
```

```python
data_sim['Gender'].value_counts().plot(kind = "bar")
```

```python
data_sim[data_sim['Age'] < 60]
```
